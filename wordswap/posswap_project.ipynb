```python
def rmvspcchar(dct):
    a = dct.replace('[', '')
    b = a.replace(']', '')
    c = b.replace('<', '')
    d = c.replace('>', '')
    return d
#think about why this method worked on the raw string before it was tokenized.
#what are the limitations on list manipulation that made removing these characters so difficult?
#maybe you were iterating through the list wrong?
```

```python
def dictNN(dct):
    d = rmvspcchar(dct)
    dctags = d.tags
    NN = [ ]
    for a in dctags:
        if a[1] == 'NN':
            if a[0][-1] != (r'=' or '+' or '~'):
                NN.append(a[0])
    return NN
```
